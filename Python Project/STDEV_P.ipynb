```python

grp = ['50-59' , '60-69' ,  '70-79', '80-89', '90-100']
X = [3 , 5, 9, 12, 8]

group =pd.DataFrame({
    'class_' : grp , 
    'X' : X 
})

group['class_']
```

```python
ungroup['mean_x'] = ungroup['X'].mean() 

ungroup['X-x'] = ungroup['X'] - ungroup['X'].mean()

ungroup['(X-x)^2'] = ungroup['X-x'].pow(2)
# age_sums = age_sum.item() 
ungroup['sum(X-x^2)'] = ungroup['(X-x)^2'].sum()


def show_group () : 
    mean_x = ungroup['X'].mean() 
    print(f"mean of X == {mean}")
    sum_pow2 = ungroup['(X-x)^2'].sum()
    print(f"sum of (X-x)^2 == {sum_pow2}")
    N = len(ungroup['X']) 
    print(f"Sample size of {N}")
    under = sum_pow2 / N
    print(f"Under : {under}")
    import math
    print(f" Standard Deviation == {math.sqrt(under)}" )
# ungroup = ungroup.reindex(columns=['X', 'mean_x', 'X-x', '(X-x)^2', ])
```

```python
group[['1st', '2nd']]= group['class_'].str.split('-', expand=True)
group[['1st', '2nd']]= group[['1st' , '2nd']].astype(int)
group['midpoint'] = (group['1st'] + group['2nd']) / 2 
group['f*midp'] = group['X'] * group['midpoint'] 
group['Mean'] = group['f*midp'].sum() / group['X'].sum()
group['m-Mean'] = group['midpoint'] - group['Mean']
group['m-Mean^2'] = group['m-Mean'].pow(2)
group['sum(m-Mean^2)'] = group['m-Mean^2'].sum()
group['N-1'] = group['X'].sum() - 1 
group["std_deviation"] = group['sum(m-Mean^2)'] / group['N-1']
print(group.dtypes)

group
```
