```python
import matplotlib.pyplot as plt

# Given LaTeX expression
latex_expr = r'\frac{n}{{N-4}}x'

# New variable values
n = 3
N = 10

# Replace variables in the LaTeX expression using format()
formatted_latex_expr = latex_expr.format(n=n, N=N)

# Plot
plt.text(0.5, 0.5, formatted_latex_expr, fontsize=16, ha='center', va='center')
plt.axis('off')  # Turn off axes
plt.show()
```
