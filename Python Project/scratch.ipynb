```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
df_sheet1 = pd.read_excel('sampol.xlsx' , sheet_name='Sheet1')
df_sheet2 = pd.read_excel('sampol.xlsx' , sheet_name='Sheet2')
df_sheet1 = df_sheet1.rename(columns=lambda x: x.replace(' ', '_'))
df_sheet2 = df_sheet2.rename(columns=lambda x: x.replace(' ', '_'))
df_sheet1['DATE'] = pd.to_datetime(df_sheet1['DATE'])
df_sheet2['DATE'] = pd.to_datetime(df_sheet2['DATE'])

df_sheet1
```

```python
df_sheet1['row_num'] = df_sheet1.groupby('LOAN_ID') .cumcount() + 1
df_sheet2['row_num'] = df_sheet2.groupby('LOAN_ID') .cumcount() + 1

df_sheet1
```

```python
merg_sheet = pd.merge(df_sheet1, df_sheet2 , on=['LOAN_ID' , 'row_num'] , how='inner') 
merg_sheet


# yung wala dito sa  df_sheet1 na meron sa df_2 malalgay siya sa df_sheet1
```
