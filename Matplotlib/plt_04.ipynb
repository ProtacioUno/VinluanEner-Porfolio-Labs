```python
from matplotlib import pyplot as plt
import random
import matplotlib.colors as mcolor
import matplotx  
import numpy as np
import csv
import os
from collections import Counter
import pandas as pd
```

```python

plt.style.use("fivethirtyeight")

# slices = [60 , 40]
labels = ["Sixy", "Forty"]



def slices():
    ramram = []
    for x in range(8):
        rand  = random.randint(20, 200)
        ramram.append(rand)
    ramram = sorted(ramram)
    return ramram    

fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig', 'grape', 'honeydew']

fig, ax = plt.subplots(facecolor='black')


plt.pie(slices() , labels=fruits , wedgeprops={'edgecolor' : 'black' , 'linewidth' : 3})


for text in ax.texts:
    text.set_color('white')

plt.title("My shitness")
plt.tight_layout()
plt.show
```

```python

plt.style.use("fivethirtyeight")

# slices = [60 , 40]
# labels = ["Sixy", "Forty"]



def slices():
    ramram = []
    for x in range(5):
        rand  = random.randint(20, 200)
        ramram.append(rand)
    # ramram = sorted(ramram)
    return ramram    

statik_slices = slices()

fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig', 'grape', 'honeydew']
fruits = fruits[:-3]

# explode = [0, 0 , 0 , 0.1 ,0 ]


# explode by base on slices maxx value
explodek = []
maxx = max(statik_slices)
zero = 0 
p_one = 0.1

for xx in statik_slices:
    if xx >= maxx:
        explodek.append(p_one)
    else: 
        explodek.append(zero)
        
fig, ax = plt.subplots(facecolor='black')


plt.pie(statik_slices , labels=fruits ,
        explode=explodek,         
        shadow=False,
        startangle=90,
        autopct='%1.1f%%', 
        wedgeprops={'edgecolor' : 'black' , 'linewidth' : 3})


for text in ax.texts:
    text.set_color('white')

plt.title("My shitness")
plt.tight_layout()
plt.show
```
