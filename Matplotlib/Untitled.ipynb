```python


fb = (120, 98, 110 , 100, 95)
yt = (80, 90 , 85, 95 , 70)
ig = (50, 40 , 55, 60 , 45)
sc = (20 , 30 , 35 , 25 ,  30)

xpos = np.arange(len(fb))


```

```python
style.use('ggplot')

plt.figure(figsize=(10, 7))
# barwidht = 0.2

plt.bar(xpos, fb , color='royalblue', width=0.3, label='FB')
plt.bar(xpos, yt  , color='red', width=0.3, label='YT')
plt.bar(xpos, ig  , color=color[2], width=0.3, label='IG')
plt.bar(xpos, sc , color=color[3], width=0.3, label='SC')
plt.show()

```
